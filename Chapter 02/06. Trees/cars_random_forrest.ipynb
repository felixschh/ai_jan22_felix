```python
import pandas as pd

pd.read_csv(car_data, header=None)
```

```python
columns = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'acceptability']
```

```python
import matplotlib.pyplot as plt
for col_name in df.columns:
    plt.figure()
    plt.hist(df[col_name])
    plt.title(col_name)
```

```python
df = pd.read_csv(car_data, names=columns)
df = df.replace({'5more': 5, 'more': 5})
df = df.replace({'vhigh': 4, 'high': 3, 'med': 2, 'low': 1})
df = df.replace({'big': 3, 'small': 1})
df = df.replace({'acc': 1, 'unacc': 0, 'good': 2, 'vgood': 3})
df
```

```python
X, y = df.iloc[:, :-1], df.iloc[:, -1]
```

```python
df = pd.read_csv(car_data, names=columns)
df = df.replace({'5more': 5, 'more': 5})
df = df.replace({'vhigh': 4, 'high': 3, 'med': 2, 'low': 1})
df = df.replace({'big': 3, 'small': 1})
df = df.replace({'acc': 1, 'unacc': 0, 'good': 2, 'vgood': 3})
df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

```python
dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf = dt_clf.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15, 15))
plot_tree(dt_clf,
          filled=True,
          rounded=True,
          class_names=['unacc', 'acc', 'good', 'vgood'],
         feature_names=X.columns)
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(dt_clf, X_test, y_test, display_labels=['unacc', 'acc', 'good', 'vgood'])
```
